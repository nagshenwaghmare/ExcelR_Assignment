Using k-mean

```python
crime1=pd.read_csv('/content/crime_data.csv')
crime1
```

```python
df1=crime1.iloc[:,1:]
df1.head()
```

```python
#normalization 
#normalized_df=(df-df.min())/(df.max()-df.min())
norm_df1=(df1-df.min())/(df.max()-df.min())
norm_df1
```

```python
# Elbow curv
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(norm_df1)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(norm_df1)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime1['Clust']=x
```

Using DBSCAN


```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime3=pd.read_csv('/content/crime_data.csv')
crime3.head()
```

```python
df3=crime3.iloc[:,1:]
stscaler=StandardScaler().fit(df3)
x=stscaler.transform(df3)
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
dbscan.labels_
```

```python
crime3['clujster']=dbscan.labels_
crime3
```
